```python
import requests
import settings
import pprint
pp = pprint.PrettyPrinter(indent=2)

base = "http://api.parsely.com/v2"

def make_api_request(uri,params={}):
    key = settings.KEY
    secret = settings.SECRET
    url = base + uri
    _params = {"apikey":key, "secret":secret}
    _params.update(params)
    r = requests.get(url,params=_params)
    print r.url
    return r.json()
```

```python
# /analytics/posts for posts published in the last 3 days
analytics = make_api_request("/analytics/posts",{"pub_days":3})
pp.pprint(analytics)
```

```python
# Most popular topics for posts published in the last 3 days
analytics = make_api_request("/analytics/topics",{"pub_days":3})
pp.pprint(analytics)
```

```python
# Diving deeper, most popular posts with "Barack Obama" as a topic, which were published during the last month
analytics = make_api_request("/analytics/topic/Barack+Obama/detail",{"pub_days":30})
pp.pprint(analytics)
```

```python
# Analytics for specific posts
url = "http://www.foxnews.com/world/2013/04/02/north-korea-vows-to-restart-nuclear-facilities-amid-tensions/"
analytics = make_api_request("/analytics/post/detail",{"url":url})
pp.pprint(analytics)
```

# Searching and recommendations

```python
# simple term search
search = make_api_request("/search",{"q":" sequester cuts"})
pp.pprint(search)
```

```python
# contextual recommendations, based on a url
url = "http://video.foxnews.com/v/2179512092001/sequester-cuts-not-really-cuts-at-all/"
recs = make_api_request("/related",{"url":url})
pp.pprint(recs)
```

```python
# personalized recommendations
# 2-step process

# 1. tell us the user has viewed a post
uuid= "foxnews-sample-uuid-1"
url = "http://www.foxnews.com/opinion/2013/03/26/rewarding-hate-intolerance-and-broken-promises-with-taxpayer-dollars/"
profile = make_api_request("/profile",{"uuid":uuid,"url":url})
pp.pprint(profile)
```

```python
# personalized recommendations
# 2-step process

# 2. ask for recommendations for this user
recs = make_api_request("/related",{"uuid":uuid})
pp.pprint(recs)
```
